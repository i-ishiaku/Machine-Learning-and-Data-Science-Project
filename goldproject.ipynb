```python
df = pd.read_csv('annual_gold_rate.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['CNY'] = df['CNY'].fillna(df['CNY'].dropna().mean())
```

```python
df.head()
```

```python
df.columns
```

```python
date =df['Date']
usd =df['USD']
eur =df['EUR']
gbp =df['GBP']
inr =df['INR']
aed =df['AED']
cny =df['CNY']
```

```python
fig,axes = plt.subplots(2,3, figsize=(20,10))
axes[0][0].bar(date,usd,color='pink')
axes[0][0].set_xlabel('YEAR',color='green',fontsize=20)
axes[0][0].set_ylabel('USD GOLD RATE',color='green',fontsize=20)
axes[0][0].set_title('IMPACT OF GOLD RATE',color='green',fontsize=15)

axes[0][1].bar(date,eur,color='orange')
axes[0][1].set_xlabel('YEAR',color='green',fontsize=20)
axes[0][1].set_ylabel('EUR GOLD RATE',color='green',fontsize=20)

axes[0][2].bar(date,gbp,color='gold')
axes[0][2].set_xlabel('YEAR',color='green',fontsize=20)
axes[0][2].set_ylabel('GBP GOLD RATE',color='green',fontsize=20)

axes[1][0].bar(date,inr,color='blue')
axes[1][0].set_xlabel('YEAR',color='green',fontsize=20)
axes[1][0].set_ylabel('INR GOLD RATE',color='green',fontsize=20)

axes[1][1].bar(date,aed,color='green')
axes[1][1].set_xlabel('YEAR',color='green',fontsize=20)
axes[1][1].set_ylabel('AED GOLD RATE',color='green',fontsize=20)

axes[1][2].bar(date,cny,color='grey')
axes[1][2].set_xlabel('YEAR',color='green',fontsize=20)
axes[1][2].set_ylabel('CNY GOLD RATE',color='green',fontsize=20)

for label in axes[0][0].get_xticklabels():
    label.set_color('crimson')
for label in axes[0][0].get_yticklabels():
    label.set_color('crimson')
```
